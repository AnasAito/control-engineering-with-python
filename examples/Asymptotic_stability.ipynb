## How can we quantify this using linear algebra ? 
- to proove the asymptoric stability of an equilibrium point we start by linearizing our system 
### Case 1 : $θ=0$
- recall that we have a differential equation that model our system dynamics 
 $$ml^{2}\ddot{θ}+b\dot{θ}+mglsin(θ)=0$$
- linearizing the system given our equilibrium state (See lectures) output   this sumplified linear system  : 
$$\dot x = Ax$$
where : $A = \begin{pmatrix} 0 & 1  \\ \frac{-g}{l} & \frac{-b}{ml^{2}} \end{pmatrix}$
-  to verify the asymptotic stability we just check the matix A eigen values 
- a negative real part would mean that this equilibrium state is asymptotic stable ! 
- to do this we will use **numpy.linalg module**

```python
import numpy.linalg as linalg 
b=0.5;m=1.0;l=1.0;g=9.81
A=np.array([[0,1],[-g/l,-b/(m*l**2)]])
eign_vals,_=linalg.eig(A)
eign_vals
```
